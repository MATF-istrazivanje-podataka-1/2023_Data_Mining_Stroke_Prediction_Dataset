```python
X= pd.read_csv('../../preprocessing/X_preprocessed1.csv')
y=  pd.read_csv('../../preprocessing/Y_preprocessed1.csv')
feature_names = X.columns
y=y['stroke']
```

```python
normalScaler = StandardScaler()
normalScaler.fit(X)
StandardScaler()

X_normalized = pd.DataFrame(normalScaler.transform(X))
clustering_a=[]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)
```

```python
X_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
```

```python
silhouettes = []
sse_scores = []
for i in range(2, 10):
    bkmeans = BisectingKMeans(n_clusters=i,  bisecting_strategy='biggest_inertia', random_state=48,init='k-means++')
    bkmeans.fit(X_normalized)
    labels = bkmeans.predict(X_normalized)
    for j in range (i):
        plt.title(f'2D: k={i} ')  
        plt.scatter(X_pca["PC1"], X_pca["PC2"], c=labels)
    silhouettes.append(silhouette_score(X_normalized, labels))
    sse_scores.append(round(bkmeans.inertia_, 4))
    plt.show()
    
```

```python
fig = plt.figure(figsize=(12, 5))

x = [f'{i}' for i in range(2,10)]

fig.add_subplot(1, 2, 1)
y1 = sse_scores
plt.plot(x, y1, marker='o')
plt.title('SSE za razlicite k')

fig.add_subplot(1, 2, 2)
y2 = silhouettes
plt.plot(x, y2, marker='o')
plt.title('Silueta za razlicite k')
```

```python
Bkm2 =  BisectingKMeans(n_clusters=2,  bisecting_strategy='biggest_inertia', random_state=48,init='k-means++')
Bkm2.fit(X_normalized)
```

```python
X_pca['labels']= Bkm2.labels_
scatter = plt.scatter(X_pca["PC1"], X_pca["PC2"],c=Bkm2.labels_)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('{} clusters'.format("Bisecting Kmeans, n_clusters = 2"))

plt.show()
```
