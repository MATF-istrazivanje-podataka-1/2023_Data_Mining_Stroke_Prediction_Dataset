```python
X= pd.read_csv('../../preprocessing/X_preprocessed1.csv')
y=  pd.read_csv('../../preprocessing/Y_preprocessed1.csv')
feature_names = X.columns
y=y['stroke']
```

```python
normalScaler = StandardScaler()
normalScaler.fit(X)
StandardScaler()

X_normalized = pd.DataFrame(normalScaler.transform(X))
clustering_a=[]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)
```

```python
X_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
k = 2
linkages = ['single', 'complete', 'average']
fig = plt.figure(figsize=(15,5))
for idx, linkage in enumerate(linkages):
    model = AgglomerativeClustering(n_clusters=k, linkage=linkage)
    model.fit(X_normalized)
    
    fig.add_subplot(1, len(linkages), idx+1)
    
    for cluster_label in range(k):
        cluster = X_pca[model.labels_ == cluster_label]
        plt.scatter(cluster['PC1'], cluster['PC2'])
    plt.title(f'l={linkage}, sil={silhouette_score(X_normalized, model.labels_)} , k=2')
   
    plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z = linkage(X_normalized, method='average')
Z
```

```python
X_normalized = X_normalized.set_index(y)
```

```python
X_normalized.index
```

```python
_ = dendrogram(Z, labels=X_normalized.index, leaf_rotation=90)
```

```python
silhouettes = []
linkages = [ 'average']
fig = plt.figure(figsize=(15,5))
for k in range(2,10):
    for idx, linkage in enumerate(linkages):
        model = AgglomerativeClustering(n_clusters=k, linkage=linkage)
        model.fit(X_normalized)
    
        fig.add_subplot(1, len(linkages), idx+1)
    
        for cluster_label in range(k):
            cluster = X_pca[model.labels_ == cluster_label]
            plt.scatter(cluster['PC1'], cluster['PC2'])
        plt.title(f'l={linkage}, sil={silhouette_score(X_normalized, model.labels_)} , k={k}')
        silhouettes.append(silhouette_score(X_normalized, model.labels_))
        plt.show()
       
```

```python
def clustering_accuracy(real, predicted):
    total = len(real)
    correct = sum(a == b for a, b in zip(real, predicted))
    score = (correct / total) * 100
    return score
clustering_acc=[]
```

```python
fig = plt.figure(figsize=(12, 5))
x = [f'{i}' for i in range(2,10)]
fig.add_subplot(1, 2, 2)
y2 = silhouettes
plt.plot(x, y2, marker='o')
plt.title('Silueta za razlicite k')

plt.show()
```

```python
model = AgglomerativeClustering(n_clusters=2, linkage='average', compute_distances=True)
model.fit(X_normalized)
```

```python
clustering_accuracy(y, model.labels_)
```
